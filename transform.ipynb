```python
import pandas as pd
```

```python
games_df = pd.read_json("games.json")
```

```python
games_df = games_df.transpose()
games_df.head()
```

```python
games_df.dtypes
```

```python
col = ['name', 'release_date', 'price', 'detailed_description', 'short_description', 'about_the_game', 'categories', 'genres', 'recommendations', 'metacritic_score', 'windows', 'mac', 'linux']
t_games_df = games_df.loc[:, col]
t_games_df.head()
```

```python
t_games_df = pd.get_dummies(t_games_df, columns=['windows', 'mac', 'linux'])
t_games_df = t_games_df.drop(columns=['windows_False', 'mac_False', 'linux_False'])
t_games_df.head()
```

```python
dupes_df = t_games_df[t_games_df.duplicated('name')]
dupes_df
```

```python
clean_df = t_games_df.drop_duplicates(subset='name')
clean_df
```

```python
clean_df['release_date'] = pd.to_datetime(clean_df['release_date'])
clean_df
```

```python
clean_df.insert(0, 'appid', clean_df.index)
clean_df['metacritic_score'] = pd.to_numeric(clean_df['metacritic_score'])
clean_df['recommendations'] = pd.to_numeric(clean_df['recommendations'])
clean_df['price'] = pd.to_numeric(clean_df['price'])
clean_df.dropna(axis=0, inplace=True, how='any')
clean_df
```

```python
clean_df.isna().sum()
```

```python
clean_df.dtypes
```

```python
tags_df = pd.read_csv('steamspy_tag_data.csv')
tags_df
```

```python
tags_df.isna().sum()
```

```python
tags_df[tags_df.duplicated('appid')]
```

```python
combined_df = tags_df.merge(clean_df, how='left', on='appid')
combined_df
```

```python
combined_df.isna().sum()
```

```python
combined_df[combined_df.duplicated('appid')].sum()
```

```python
combined_df = combined_df.dropna()
combined_df
```

```python
combined_df.to_csv('game_data.csv', index=False)
```
